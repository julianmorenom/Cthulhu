### Training 

Run the training script.

* Change the data folder to the image folder on google drive, you can copy the path in the file browser with right click.
* Also change the output folder to some place on your GDrive.
* For different resolutions you can pick these pretrained networks by setting the resume parameter: ffhq1024, ffhq512, ffhq256
* If you want to resume from your own model you need to specify a pkl file from your out folder, e.g. /content/drive/MyDrive/MachineLearning/stylegan-out/00000-flickrMarineBiology_1024-mirror-auto1-kimg20-ada-resumeffhq1024/network-snapshot-000008.pkl
* You can check the progress in the out folder. Every 20-30 minutes you should see a fakes0000??.png file with generated samples. 

```
!python train.py \
    --kimg=200 \
    --metrics=none \
    --outdir=/content/drive/MyDrive/MachineLearning/stylegan-out \
    --data=/content/drive/MyDrive/MachineLearning/flickrMarineBiology_1024 \
    --mirror=true \
    --aug=ada \
    --cfg=auto \
    --snap=1 \
    --resume=/content/drive/MyDrive/MachineLearning/stylegan-out/00007-flickrMarineBiology_1024-mirror-auto1-kimg500-ada-resumecustom/network-snapshot-000184.pkl \
  
```
